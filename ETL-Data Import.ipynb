## Import Traffic

```python
traffic_yosemite = pd.read_csv('TrafficCounts/YosemiteTraffic.csv')
# traffic_yosemite.head()
traffic_yosemite_group = traffic_yosemite.groupby(["Year", "GroupDescription"])
traffic_yosemite_group = traffic_yosemite_group.sum()
traffic_yosemite_group = traffic_yosemite_group.sort_values("Year", ascending = False)
traffic_yosemite_group.head()
```

```python
traffic_yellowstone = pd.read_csv('TrafficCounts/YellowstoneTraffic.csv')
# traffic_yellowstone.head()
traffic_yellowstone_group = traffic_yellowstone.groupby(["Year", "GroupDescription"])
traffic_yellowstone_group = traffic_yellowstone_group.sum()
traffic_yellowstone_group = traffic_yellowstone_group.sort_values("Year", ascending = False)
traffic_yellowstone_group.head()
```
